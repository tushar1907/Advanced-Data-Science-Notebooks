```python
from __future__ import print_function
import urllib3 #  pip install -i https://pypi.anaconda.org/pypi/simple urllib3
import unicodedata
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
warnings.filterwarnings('ignore')
import csv
    
    
def  getstatus(url):    
     http = urllib3.PoolManager()
     response = http.request('GET',url)
     return response
```

```python
html = urllib.request.urlopen("https://www.google.com")
soup = BeautifulSoup(html, 'lxml')
tags = soup.find_all('a')
```

```python
data = []
counter = 0
spamwriter = csv.writer(open('Data.csv',"w", encoding="utf-8"),quotechar=" ")
spamwriter.writerow(["Links"  ,  "Texts"])
 
for t in tags:
    if(str(t['href']).startswith('http')):
        if(getstatus(t['href']).status == 200):
            data.insert(0,str(t['href'])+'\t,"{}"'.format(str(t.string)))
            spamwriter.writerow([data[0]])
            print(data[0])
            counter+=1  

#links and texts are saved in one excel sheet eg. Data.csv
```
